```r
library('tidyverse')

options(repr.plot.res = 300, repr.plot.height = 3, repr.plot.width = 4)
```

```r
df_turbine = read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv", col_types = cols())

df_turbine %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
p = ggplot(df_turbine, aes(y = province_territory %>% fct_infreq %>% fct_rev))

p = p + labs(y = "Province Territory")

p = p + geom_bar()

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
p = ggplot(df_turbine, aes(y = manufacturer %>% fct_infreq %>% fct_lump(n=5) %>% fct_rev))

p = p + geom_bar()

p = p + labs(y = "Manufacturer")
           
p
```
